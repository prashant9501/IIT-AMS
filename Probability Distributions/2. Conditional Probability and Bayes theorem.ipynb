### CONDITIONAL PROBABILITY

- Prob for an event A occuring, knowing (given) that event B has already occured.
- P(A|B) >>> read as ....> Prob. of A given B

P(A|B) = P(A and B)/P(B)   >>> ratio of the joint prob. and the
marginal prob. of the event which has already occured
